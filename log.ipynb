```python
import cv2
import matplotlib.pyplot as plt
```

```python
#reading the image
img=cv2.imread("hand.jpg")
plt.imshow(img)
```

```python
#blur the image
blur_img=cv2.GaussianBlur(img,(3,3),0)
plt.imshow(blur_img)
```

```python
#apply the laplacian operator 
lap_img=cv2.Laplacian(blur_img,cv2.CV_64F)
plt.imshow(lap_img)
```

```python
laplacian1=lap_img/lap_img.max()
plt.imshow(laplacian1,cmap='gray')
```
