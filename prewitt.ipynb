```python
#prewitt edge detection
import cv2
import numpy as np
from matplotlib import pyplot as plt
img=cv2.imread("hand.jpg")
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
#original image
plt.imshow(img)
```

```python
#blurring the image
blur_img=cv2.GaussianBlur(img,(3,3),1000)
plt.imshow(blur_img)
```

```python
#kernel array
kernelx=np.array([[1,1,1],[0,0,0],[-1,-1,-1]])
kernely=np.array([[-1,0,1],[-1,0,1],[-1,0,1]])
```

```python
#prewitt on x axis
prewitt_imgx=cv2.filter2D(blur_img,-1,kernelx)
plt.imshow(prewitt_imgx)
```

```python
#prewitt on y axis
prewitt_imgy=cv2.filter2D(blur_img,-1,kernely)
plt.imshow(prewitt_imgy)
```
